```python
n = int(input('enter an integer: '))

#Universe = integers
#B = numbers divisble by 2
#C = negative numbers
#A = positive numbers divisible by 2

if (n%2 != 0) and (n > 0): 
    print(f'{n} is a positive odd number')
elif (n%2 == 0) and (n <= 0):
    print(f'{n} is divisible by 2 but not positive')
elif (n%2 == 0) and (n < 0):
    print(f'{n} a negative even number')
elif (n%2 != 0) and (n < 0):
    print(f"{n} is an odd negative number")
else:
    print(f'{n} is a positive even number')
```
